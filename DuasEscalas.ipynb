```
#
#######################################################################
# Objetivo
# Usar graficos com duas escalas, como opção a mostrar
# simplesmente uma tabela.
#
# https://matplotlib.org/examples/api/two_scales.html
#
#######################################################################
#

#
#######################################################################
# Importando as Bibliotecas
#######################################################################
#
import numpy as np
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()
t = np.arange(0.01, 10.0, 0.01)
s1 = np.exp(t)
ax1.plot(t, s1, 'b-')
ax1.set_xlabel('time (s)')

#
#######################################################################
# Desenhando o primeiro eixo
#######################################################################
#
ax1.set_ylabel('exp', color='b')
ax1.tick_params('y', colors='b')

#
#######################################################################
# Desenhando o segundo eixo
#######################################################################
#
ax2 = ax1.twinx()
s2 = np.sin(2 * np.pi * t)
ax2.plot(t, s2, 'r.')
ax2.set_ylabel('sin', color='r')
ax2.tick_params('y', colors='r')

#
#######################################################################
# Juntando os graficos e mostrando
#######################################################################
#
fig.tight_layout()
plt.show()
```

```
#
#######################################################################
# Objetivo
# Usar mapas de calor para apresentar numeros, como opção a mostrar
# simplesmente uma tabela.
#
# https://stackoverflow.com/questions/12286607/making-heatmap-from-pandas-dataframe
# https://indianaiproduction.com/seaborn-heatmap/
# https://likegeeks.com/seaborn-heatmap-tutorial/
# https://indianaiproduction.com/seaborn-heatmap/
#
#######################################################################
#


#
#######################################################################
# Importando as Bibliotecas
#######################################################################
#
import seaborn as sns # for data visualization
import pandas as pd # for data analysis
import numpy as np # for numeric calculation
import matplotlib.pyplot as plt # for data visualization

#
#######################################################################
# Definindo os Pontos do Gráfico
#######################################################################
#
Index= ['aaa', 'bbb', 'ccc', 'ddd', 'eee']
Cols = ['A', 'B', 'C', 'D']
df = DataFrame(abs(np.random.randn(5, 4)), index=Index, columns=Cols)

#
#######################################################################
# Plotando o Gráfico
#######################################################################
#
plt.figure(figsize=(10,9))
sns.set(font_scale=1.5)
sns.heatmap(df, cmap='RdYlGn_r', linewidths=0.5, fmt=".2f", annot=True)
plt.xlabel("Values on X axis",fontweight="bold")
plt.ylabel("Values on Y axis",fontweight="bold")
plt.title("Título do Gráfico\n",fontweight="bold")
plt.show()
```

```
#
#######################################################################
# Definindo os Pontos do Gráfico
#######################################################################
#

Index= ['aaa', 'bbb', 'ccc', 'ddd', 'eee']
Cols = ['A', 'B', 'C', 'D']
df = DataFrame(abs(np.random.randn(5, 4)), index=Index, columns=Cols)

#
#######################################################################
# Plotando o Gráfico
#######################################################################
#
plt.figure(figsize=(10,9))
sns.set(font_scale=1.5)
sns.heatmap(df, annot=True, cmap='RdYlGn_r', linewidths=0.5, fmt=".2f", cbar_kws={'label': 'Escala Utilizada', 'orientation': 'horizontal'})
plt.xlabel("Values on X axis",fontweight="bold")
plt.ylabel("Values on Y axis",fontweight="bold")
plt.title("Título do Gráfico\n",fontweight="bold")
plt.show()
```
